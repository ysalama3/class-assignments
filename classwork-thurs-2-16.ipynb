```python
def makeBirdDict(filename):
    #this function opens the data file from Thomas et al and extracts the species and family as a dictionary.
    birddict= {}
    #open the file
    fobj = open(filename,"r", encoding = "ISO-8859-15")
    #read the file
    all_records = fobj.readlines()[1:]
    #split the lines on comma 
    #stic species and families into bird dict
    for line in all_records:
        species = line.split(",")[3]
        families = line.split(",")[8]
        birddict[species] = families
        
    return birddict
```
